```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# creates path to chrome 
browser = Browser('chrome')
```

```python
url = "https://www.tn.gov/tdfi/tdfi-how-do-i/info/mortgage-rate-of-interest/max-effective-interest-rate-history.html"
```

```python
# activates path to chrome (python has control of page)
browser.visit(url)
```

```python
html = browser.html 
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
html
```

```python
link = soup.find_all("div", class_ = "tn-rte")
```

```python
link
```

```python
link[0].find_all('p')[2]
```

```python
for x in link[0].find_all("p")[2]:
    print(x.text)
```

```python
if link:
    # Grab the last <p> tag's content
    last_p_content = link[0].find_all('p')[2].get_text()
    print(last_p_content)
```

```python
# Split by newline to get each row
rows = last_p_content.split('\n')

# List to store the data
data = []

# Regex pattern to extract month, year, and rate
pattern = re.compile(r"([A-Za-z]+,\s\d{4})\s+(\d+\.\d+)")

for row in rows:
    match = pattern.search(row)
    if match:
        month_year = match.group(1)
        rate = match.group(2)
        data.append({"Month_Year": month_year, "Rate": rate})

# Convert to DataFrame
df_rates = pd.DataFrame(data)
```
